```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
comb_samp.iloc[:,1:]= scaler.fit_transform(comb_samp.iloc[:,1:])
```

```python
comb_samp.head()
```

```python
X= comb_samp.iloc[:,1:]
Y = comb_samp['attacktype']
```

```python
Y.index
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import CategoricalNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import f1_score, classification_report
from sklearn.metrics import recall_score
```

```python
svc = SVC(kernel='sigmoid', gamma=1.0)
knc = KNeighborsClassifier(n_neighbors=5)
mnb = MultinomialNB(alpha=0.2)
cnb = CategoricalNB()
dtc = DecisionTreeClassifier(min_samples_split=2, random_state=111)
lrc = LogisticRegression(solver='liblinear', penalty='l1')
rfc = RandomForestClassifier(n_estimators=50, random_state=111)
abc = AdaBoostClassifier(n_estimators=62, random_state=111)
bc = BaggingClassifier(n_estimators=9, random_state=111)
etc = ExtraTreesClassifier(n_estimators=9, random_state=111)
dnn = MLPClassifier(solver='sgd', alpha=1e-5, learning_rate = 'adaptive', activation = 'logistic')
```

```python
clfs = {'SVC' : svc,'KN' : knc, 'MNB':mnb, 'DT': dtc, 'LR': lrc, 'RF': rfc, 'AdaBoost': abc, 'BgC': bc, 'ETC': etc, 'DNN':dnn}
```

```python
def train_classifier(clf, feature_train, labels_train):    
    clf.fit(feature_train, labels_train)
```

```python
# function to predict using classifiers
def predict_labels(clf, features):
    return (clf.predict(features))
```

```python
X_test.index
```

```python
pred_scores = []
for k,v in clfs.items():
    train_classifier(v, X_train, y_train)
    pred = predict_labels(v,X_test)
    print(k,classification_report(y_test,pred))
    print(k,'Specificity',recall_score(y_test, pred, pos_label='other_type_pain'))
```
